```python


# Define MiDaS input transformation
transform = Compose([
    Resize((256, 256)),  # Resize for MiDaS model
    ToTensor(),
    Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Start video capture
cap = cv2.VideoCapture(0)  # Change to 1 if needed

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read frame.")
        break

    # YOLO Object Detection
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = yolo_model(img_rgb)  # YOLO inference

    # Check if results contain any detections
    boxes = results[0].boxes  # Extract bounding boxes
    output_frame = frame.copy()  # Frame to display

    # Prepare MiDaS input
    input_batch = transform(Image.fromarray(img_rgb)).unsqueeze(0).to('cpu')

    with torch.no_grad():
        depth_map = midas(input_batch)  # Get MiDaS depth map
        depth_map = torch.nn.functional.interpolate(
            depth_map.unsqueeze(1),
            size=frame.shape[:2],
            mode="bicubic",
            align_corners=False
        ).squeeze().cpu().numpy()

    # Normalize and apply color map to depth map for visualization
    depth_map_normalized = (depth_map / depth_map.max()) * 255
    depth_map_normalized = depth_map_normalized.astype(np.uint8)
    depth_map_color = cv2.applyColorMap(depth_map_normalized, cv2.COLORMAP_INFERNO)

    # Process bounding boxes and display depth for each detected object
    for box in boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0].cpu().numpy())  # Bounding box coordinates

        # Extract depth values for the object's bounding box region
        object_depth_map = depth_map[y1:y2, x1:x2]

        # Calculate the median depth
        median_depth = np.median(object_depth_map) if object_depth_map.size > 0 else 0

        # Draw bounding box and depth information
        cv2.rectangle(output_frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        text_position = (x1, y1 - 10) if y1 > 20 else (x1, y2 + 20)
        cv2.putText(output_frame, f"Depth: {median_depth:.2f}", text_position, cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # Display YOLO detections and depth map
    cv2.imshow('YOLO Real-Time Detection', output_frame)
    cv2.imshow('Depth Map', depth_map_color)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close windows
cap.release()
cv2.destroyAllWindows()
```
